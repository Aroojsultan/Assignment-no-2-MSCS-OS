```python
#6.11 program
print("\n\t%%% This is cities program(6.11) %%%\n")
cities_arooj = {'Multan':{'Country':'Pakistan', 'Population':'ten lac', 'Facts':' city of saints.'},
'Lahore':{'Country':'Pakistan', 'Population':'one billion', 'Facts':' a major centre of Qawwali music.'}, 'Islamabad':{'Country':'Pakistan', 'Population':'ten billion', 'Facts': ' capital of Pakistan.'}}
for city, city_info in cities_arooj.items():
    Country = city_info['Country']
    Population= city_info['Population']
    Facts= city_info['Facts']
    print("\n"+ city + " is in " + Country)
    print("\n"+ city + " has population of " + Population)
    print("\n"+ city + " major fact is " + Facts+ '\n')
```
